### 1. Dismissal types with count.

```python
dismissal_types_count = df[['dismissal_kind']].value_counts()
dismissal_types_count
```

### 2. List of 10 fielders with the greatest number of catches.

```python
fielder = df.groupby(['dismissal_kind'])
no_of_catches = fielder['fielder'].value_counts()[['caught']].head(10)
no_of_catches
```

### 3. List of top 10 highest runs scored by batsman.

```python
highest_runs = df.groupby(['batsman'])['batsman_runs'].sum().sort_values(ascending=False).head(10)
highest_runs
```

### 4. List of top 10 bowlers with highest wickets taken.

```python
most_number_wickets = df.groupby(['bowler'])[['dismissal_kind']].apply(lambda x: x.isin(
    ['caught', 'lbw', 'bowled', 'stumped', 'caught and bowled', 'hit wicket']).sum())
most_number_wickets = most_number_wickets.sort_values(by='dismissal_kind', ascending=False)[:10]
most_number_wickets.rename(columns={'dismissal_kind': 'No. of Wickets'}, inplace=True)
most_number_wickets
```

### 5. List of 10 bowlers who bowled most number of no balls.

```python
no_ball = (df['extras_type'] == 'noballs')
no_ball = df[no_ball].groupby(['extras_type'])
no_ball = no_ball['bowler'].value_counts().head(10).sort_values(ascending=False)
no_ball
```

### 6. List of 10 players with most Man of the Match titles.

```python
titles = df1[['player_of_match']].value_counts().head(10)
titles
```

### 7. List of teams and the number of sixes (6s) scored by each team.

```python
teams = df.groupby(['batting_team'])
no_of_sixes = teams[['batsman_runs']].apply(lambda x: x.isin([6]).sum())
no_of_sixes.rename(columns = {'batsman_runs': 'No. of sixes'}, inplace = True)
no_of_sixes
```

### 8. Total runs scored by the batsman for each team he played.

```python
batsman_team = df.groupby(['batsman', 'batting_team'])
batsman_team = batsman_team[['batsman_runs']].sum()
batsman_team
```

### 9. Matches won by each team in each season. [including eliminator and finals]

```python
df1['season'] = pd.DatetimeIndex(df1['date']).year
win = df1.groupby(['season'])
win = win['winner'].value_counts()
win
```

### 11. Teams having most Man of the Match.

### 12. Man of the match in each season and runs scored by man of the match.

```python
join_df = pd.merge(left=df, right=df1, left_on='match_id', right_on='match_id')
```

```python
join_df.head()
```

```python
player = (join_df['player_of_match'] == join_df['batsman'])
mom_runs = join_df[player].groupby(['season', 'match_id', 'winner', 'player_of_match'])
mom_runs = mom_runs[['batsman_runs']].sum()
mom_runs.rename(columns={'batsman_runs': 'Runs by man of the match'}, inplace = True)
mom_runs
```

### 13. Highest score by individual batsman.

```python
join_df1 = pd.merge(left=join_df, right=df2, left_on='venue_id', right_on='venue_id')
join_df1.head()
```

```python
highest_score = join_df1.groupby(['season', 'venue', 'match_id', 'batsman', 'batting_team'])[["batsman_runs"]].sum()
highest_score = highest_score.sort_values(by='batsman_runs', ascending=False)[:10]
highest_score
```

### 14. Wickets taken by Man of the Match.

```python
player = (join_df['player_of_match'] == join_df['bowler'])
mom = join_df[player].groupby(['match_id', 'winner', 'player_of_match'])
mom_wickets = mom[['dismissal_kind']].apply(lambda x: x.isin(
    ['caught', 'lbw', 'bowled', 'stumped', 'caught and bowled', 'hit wicket']).sum()
    ).sort_values(by='dismissal_kind', ascending=False)
mom_wickets.rename(columns={'dismissal_kind': 'Wickets Taken by M.O.M'}, inplace = True)
mom_wickets
```

#### If you look at the data there are players with 0 wickets, this is because they have bowled 1 or few overs and recieved M.O.M for their batting performance.

### 15. List of team and boundary scored when the team won the toss and decided to bat first. (boundary includes 4s and 6s)

```python
toss = (join_df['toss_decision'] == 'bat')
teams = join_df[toss].groupby(['toss_winner', 'toss_decision'])
boundaries = teams[['batsman_runs']].apply(lambda x: x.isin([4,6]).sum())
boundaries.rename(columns={'batsman_runs': 'No. of Boundaries'}, inplace=True)
boundaries
```

### 16. Average runs scored per match in powerplay.

```python
run = (join_df1['overs'] >= 0) & (join_df1['overs'] <= 5)
avg_runs = join_df1[run].groupby(['season', 'city', 'venue', 'match_id', 'batting_team'])
avg_runs = np.round(avg_runs[['total_runs']].sum().apply(lambda x: x/6),2)
avg_runs.rename(columns = {'total_runs': 'Run Rate in Powerplay'}, inplace=True)
avg_runs
```

### 17. Batsman runs of respective teams in each season.

```python
runs = join_df1.groupby(['season', 'batting_team', 'batsman'])
runs = runs[['batsman_runs']].sum()
runs
```

### 18. Highest number of sixes in a single match.

```python
six = join_df1.groupby(['season', 'venue', 'match_id'])
most_sixes = six[['batsman_runs']].apply(lambda x: x.isin([6]).sum()).sort_values(by='batsman_runs', ascending=False)
most_sixes.rename(columns = {'batsman_runs': 'Total Number of 6\'s'}, inplace=True)
most_sixes.head(10)
```
